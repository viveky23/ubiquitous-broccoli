```python
import pickle
import pandas as pd
```

```python
file_name_r1 = "/Users/sudarshinityagi/PycharmProjects/CCM/ubiquitous-broccoli/Human Experiments/Data/RL_trajectories_250.pickle"
file_obj_r1 = open(file_name_r1, 'rb')
trajectories_1 = pickle.load(file_obj_r1)
trajectories_1 = [list(ele) for ele in trajectories_1]

file_name_r2 = "/Users/sudarshinityagi/PycharmProjects/CCM/ubiquitous-broccoli/Human Experiments/Data/RL_trajectories_500.pickle"
file_obj_r2 = open(file_name_r2, 'rb')
trajectories_2 = pickle.load(file_obj_r2)
trajectories_2 = [list(ele) for ele in trajectories_2]

```

```python
data_1 = []
for idxt, trajectory in enumerate(trajectories_1):
    min_price_index = trajectory.index(min(trajectory))
    trajectory_number = idxt
    route = 1
    for idxd, day in enumerate(trajectory):
        if(idxd <= min_price_index): 
            
            days_left = len(trajectory) - idxd
            price = day
            if(idxd == min_price_index):
                label = 1
            else:
                label = 0
            data_1.append((trajectory_number, days_left, price, route, label))
        else:
            break
```

```python
data_2 = []
for idxt, trajectory in enumerate(trajectories_2):
    min_price_index = trajectory.index(min(trajectory))
    trajectory_number = idxt
    route = 2
    for idxd, day in enumerate(trajectory):
        if(idxd <= min_price_index): 
            
            days_left = len(trajectory) - idxd
            price = day
            if(idxd == min_price_index):
                label = 1
            else:
                label = 0
            data_2.append((trajectory_number, days_left, price, route, label))
        else:
            break
```

```python
data_1.extend(data_2)
```

```python
dataframe_object = pd.DataFrame(data_1, columns = ['trajectory_number','days_left' , 'price' , 'route', 'label'])
```

```python
Y = dataframe_object.label
```

```python
dataframe_object = dataframe_object.drop("label", axis=1)
```

```python
X = dataframe_object
```

## Decision Tree Classification

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
clf_gini = tree.DecisionTreeClassifier(criterion="gini").fit(X_train, Y_train)
Y_predicted_dt = clf_gini.predict(X_test)

print(clf_gini.score(X_test, Y_test))
```

```python
predictions = list(clf_gini.predict(X_test))
X_test['label'] = predictions
```
